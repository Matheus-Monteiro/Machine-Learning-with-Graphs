```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
def from_matrix_to_networkx(adj):
    N = len(adj[0])
    G = nx.Graph()

    for i in range(0, N):
        for j in range(0, N):
            if adj[i][j] == 1:
                G.add_edge(i, j)
    
    return G
```

```python
np_load_old = np.load
np.load = lambda *a,**k: np_load_old(*a, allow_pickle=True, **k)
with open('graphs.npy', 'rb') as f:
    g = np.load(f)
```
